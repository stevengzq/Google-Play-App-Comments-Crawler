```python
dic_data = {}
lst_data = []
lst_index = []

# æŒ‰ç…§åˆ—æ¥è¿­ä»£éå†å„ä¸ªæ ‡ç­¾ä¸‹çš„å…³é”®è¯
for i in lexicon.iteritems():
    c = 0   
    # åˆ é™¤æ¯ä¸ªå…³é”®è¯Seriesä¸‹çš„ç©ºæ¡ç›®
    for j in i[1].dropna():       
        # åˆ©ç”¨ä¸€ä¸ªboolean seriesè¿”å›df2ä¸­åŒ…å«è¯¥å…³é”®è¯çš„æ¡ç›®
        df3 = df2[df2['Comments'].str.contains(j, case=False, regex=True)]
        if c == 0:        
            # åˆå§‹åŒ–dfï¼Œ ç”¨äºè®°å½•è¯¥æ ‡ç­¾ä¸‹æ‰€æœ‰å…³é”®è¯æ¡ç›®è¯„è®ºçš„å¹¶é›†ï¼ˆå»é‡ï¼‰
            df = df3
        else:           
            # åˆå¹¶å½“å‰å…³é”®è¯è¯„è®ºå’Œä¸Šä¸€æ¡å…³é”®è¯è¯„è®ºå¹¶å»é‡ï¼ˆæ±‚å¹¶é›†ï¼‰
            df = pd.merge(df3, df, how='outer')
        c += 1
        
    print('{} : {:.2%}'.format(i[0] , len(df)/len(df2)))
    lst_data.append(len(df)/len(df2))
    lst_index.append(i[0])
    
    dic_data[i[0]] = df

pd.DataFrame(data=lst_data, index=lst_index)    
```
